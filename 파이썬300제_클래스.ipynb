```python
# 파이썬 초급 300제 261번
class Stock:
    def __init__(self, name, code):
        self.name = name
        self.code = code

    def set_name(self, name):
        self.name = name   

삼성 = Stock("삼성전자", "005930")


print(삼성.name)
print(삼성.code)
    
```

```python
a = Stock(None, None)
a.set_name("삼성전자")

print(a.name)

```
